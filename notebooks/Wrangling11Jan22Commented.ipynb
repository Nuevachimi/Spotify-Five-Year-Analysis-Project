## Wrangle genres

#### Remove quotes (evaluate string using lambda)

```python
df['track_artists_genres']
```

```python
df['track_artists_genres'] = df['track_artists_genres'].apply(lambda v: eval(v))
df['track_artists_genres']
```

### Reshape track_artists_genres

#### Create dictionary to contain oldgenre:new genre pairs

```python
genres_dict = {}
```

#### Loop though table to add to dictionary of pairs

```python
for index in genres_df.index:
    genres_dict[genres_df.iloc[index,1]] = genres_df.iloc[index,2]
```

```python
genres_dict
```

#### Create list of track_artists_genres variable

```python
track_artists_genres_list = list(df["track_artists_genres"])
track_artists_genres_list
```

#### Select the first artist's genre list only if the cell value is a list of genre lists for multiple artists, otherwise return just the single artist's genre list if there's only one.

```python
reshaped_list = [original_list[0] if type(original_list[0]) == list else original_list for original_list in track_artists_genres_list]
```

```python
df["track_artists_genres"] = reshaped_list
```

```python
df
```

#### Consolidate track_artists_genres by replacing old genres with new genres

```python
df["track_artists_genres_consolidate"] = df["track_artists_genres"].apply(lambda l: [genres_dict.get(v) for v in l])
```

```python
df["track_artists_genres_consolidate"]
```

#### Function for dummy coding genre columns

```python
def dummy_code(l, new_col):
    if new_col in l:
        return(1)
    else:
        return(0)
```

#### Dummy code genre columns creating a column for each new genre with the 0/1 presence of that genre for this track

```python
for new_col in genres_df['New Genre'].unique():
    df[new_col] = df["track_artists_genres_consolidate"].apply(lambda v: dummy_code(v, new_col))
```

```python
df
```

#### Drop original index column

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df
```

#### Export to csv

```python
df.to_csv("../data/allchartalbumtracks_RECODED.csv")
```

### Wrangle artist popularity and followers

```python
df1 = pd.read_csv("../data/allchartalbumtracks_RECODED.csv",index_col=False)
```

#### Drop old index column

```python
df1 = df1.drop("Unnamed: 0", axis=1)
```

View first popularity value

```python
df1['track_artists_popularity'][0]
```

```python
#track_artists_popularity': pd.eval, 'track_artists_genres': pd.eval, 
                            # 'track_artists_ids': pd.eval, 'track_artists_popularity': pd.eval, 
                            # 'track_artists_followers': pd.eval}
```

#### Eval for proper list format

```python
df1['track_artists_popularity'] = df1['track_artists_popularity'].apply(lambda v: eval(v))
```

```python
df1['track_artists_followers'] = df1['track_artists_followers'].apply(lambda v: eval(v))
```

#### Select only first artist info if more than one in list, else select the one artist

```python
def list_recode(v):
    if type(v) == list:
        return(v[0])
    else:
        return(v)
```

```python
df1['track_artists_popularity_first'] = df1['track_artists_popularity'].apply(lambda v: list_recode(v))
```

```python
df1['track_artists_followers_first'] = df1['track_artists_followers'].apply(lambda v: list_recode(v))
```

```python
df1
```

```python
df1['track_artists_followers']
```

#### Eval followers to return proper list format, recode as int

```python
df1['track_artists_followers'] = df1['track_artists_followers'].apply(lambda v: eval(v))
```

```python
def recode(v):
    if type(v) == list:
        return(v)
    else:
        return([int(v)])
```

```python
df1['track_artists_followers'] = df1['track_artists_followers'].apply(lambda v: recode(v))
```

#### Return sum of followers for all track artists

```python
df1['track_artists_combined_followers'] = df1['track_artists_followers'].apply(lambda v: sum(np.array(v)))
```

```python
df1['track_artists_combined_followers']
```
